```python
import pandas as pd
```

```python
df = pd.read_csv('transformer_test.csv')
df.head()
```

```python
cat_cols = list(df.select_dtypes(include=["object_"]))

df = pd.get_dummies(df, columns = cat_cols)
```

```python
new_cols = list(df)

for col in cat_cols:

    required_cols = []
    
    for i in new_cols:
        if col in i:
            required_cols.append(i)

    df.loc[:, col] = df[required_cols].idxmax(axis=1)
    df = df.drop(required_cols, axis=1)
    df.loc[:, col] = df[col].str.replace(f"{col}_", "")

    
df.head()
```
