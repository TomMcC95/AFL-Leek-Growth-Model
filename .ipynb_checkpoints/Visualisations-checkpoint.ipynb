```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import date
```

```python
summary_data_path = "F:\\Farm Data\Growth Model\Python\summary_data.xlsx"
summary_data = pd.read_excel(summary_data_path)
field_data_path = "F:\\Farm Data\Growth Model\Python\\field_summary.xlsx"
field_data = pd.read_excel(field_data_path)
```

```python
plt.rcParams['savefig.bbox'] = 'tight'
plt.rcParams["savefig.jpeg_quality"] = 75
plt.rcParams["savefig.format"] = 'jpeg'

df = summary_data.query('field == "Grotec Long 15"')
df = df[df['sample_date'].isin(['03/08/2020', '11/09/2020', '24/11/2020'])]
sns.displot(df, x="stripped_diameter", hue="sample_date", kind="kde", fill=True, bw_adjust=2, common_norm=False)
plt.axvline(x=20, color='grey')
plt.axvline(x=42, color='grey')
name = 'Grotec Long 15'
plt.title(name)
plt.savefig(name)
plt.show()

df = summary_data.query('field == "Grotec Long 15"')
df = df[df['sample_date'].isin(['24/11/2020'])]
sns.displot(df, x="stripped_diameter", kind="kde", color='g', fill=True, bw_adjust=2, cumulative=True)
plt.axvline(x=20, color='grey')
plt.axvline(x=42, color='grey')
plt.axhline(0.09, ls='--')
plt.axhline(0.885, ls='--')
name = 'Grotec Long 15 Cumulative Distribution'
plt.title(name)
plt.savefig(name, bbox_inches='tight')
plt.show()

df = summary_data.query('field == "Garretts Decoy 3"')
sns.displot(df, x="stripped_diameter", hue="sample_date", kind="kde", fill=True, bw_adjust=2, common_norm=False)
plt.axvline(x=20, color='grey')
plt.axvline(x=42, color='grey')
name = 'Garretts Decoy 3 (1)'
plt.title(name)
plt.savefig(name)
plt.show()

df = summary_data.query('field == "Garretts Decoy 3"')
df = df[df['sample_date'].isin(['19/11/2019'])]
sns.displot(df, x="stripped_diameter", kind="kde", color='g', fill=True, bw_adjust=2, common_norm=False)
plt.axvline(x=20, color='grey')
plt.axvline(x=42, color='grey')
plt.xlim(0, 70)
plt.ylim(0, 0.08)
name = 'Garretts Decoy 3'
plt.title(name)
plt.savefig(name)
plt.show()

df = summary_data.query('field == "Garretts Decoy 3"')
df = df[df['sample_date'].isin(['19/11/2019'])]
sns.displot(df, x="stripped_diameter", kind="kde", color='g', fill=True, bw_adjust=2, cumulative=True)
plt.axvline(x=20, color='grey')
plt.axvline(x=42, color='grey')
plt.axhline(0.02, ls='--')
plt.axhline(0.65, ls='--')
name = "Garretts Decoy 3 Cumulative Distribution"
plt.title(name)
plt.savefig(name)
plt.show()
```

```python
for soil in field_data.soil_type.unique():
    df = field_data
    df = field_data.query('inputs != "Baby"')
    df = df.query(f'soil_type == "{soil}"')
    sns.histplot(df, x="establishment", kde=True, stat='probability', common_norm=False, binwidth=0.05)
    plt.axvline(x=0.85, color='grey')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.title(f'Soil Type = {soil}')
    plt.savefig(f'{soil} establishment')
    plt.show()
```

```python
for soil in field_data.soil_type.unique():
    df = field_data
    df = field_data.query('inputs != "Baby"')
    df = df.query(f'soil_type == "{soil}"')
    sns.histplot(df, x="establishment", hue="method", kde=True, stat='probability', common_norm=False, binwidth=0.05)
    plt.axvline(x=0.85, color='grey')
    plt.xlim(0, 1)
    plt.ylim(0, 1.4)
    plt.title(f'Soil Type = {soil}')
    plt.savefig(f'{soil} establishment by method')
    plt.show()
```

```python
df = field_data
df = field_data.query('inputs != "Baby"')
#plt.rcParams['figure.figsize'] = (15,10)
df = df[df['variety'].isin(['Krypton', 'Chiefton', 'Shafton', 'Comanche'])]
sns.histplot(df, x="establishment", hue='variety', kde=True, stat='probability', common_norm=False, binwidth=0.05)
plt.axvline(x=0.85, color='grey')
plt.xlim(0.25,1.25)
plt.ylim(0, 0.6)
plt.savefig('variety establishment comparison')
plt.show()
```
